```python
from tkinter import *
windows = Tk()
windows.title('TK#3')

enter = Label(windows, text = 'Enter the weight in Kg')
gam = Label(windows, text = 'Gram')
pound = Label(windows, text = 'Pounds')
ounce = Label(windows, text = 'Ounce')

result_gam = Label(windows, text = 'Gram')
result_pound = Label(windows, text = 'Pounds')
result_ounce = Label(windows, text = 'Ounce')

enter_et = Entry(windows)

def convert_gam():
    result1 = float(enter_et.get()) * 1000
    result2 = float(enter_et.get()) * 2.20462262185
    result3 = float(enter_et.get()) * 35.27
    result_gam.config(text = result1)
    result_pound.config(text = result2)
    result_ounce.config(text = result3)
    
    
convert = Button(windows, text = 'Convert',command = convert_gam)

enter.grid(row=0, column=0)
gam.grid(row =1, column=0)
pound.grid(row=1, column=1)
ounce.grid(row=1, column=2)

convert.grid(row=0, column=3)

enter_et.grid(row=0, column=1)

result_gam.grid(row=2, column=0)
result_pound.grid(row=2, column=1)
result_ounce.grid(row=2, column=2)

windows.mainloop()
```
