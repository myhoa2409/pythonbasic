```python
import turtle
p=turtle.Turtle()
p.speed(30)
p.color('#020C03')
p.fillcolor("#63DC16")

p.getscreen().bgcolor("#A8F3EE")
#gọi hàm vẽ hình chữ nhật
def draw_rectangular(h,w,fillcolor = '#63DC16' ,color = '#020C03'):
    p.color(color)
    p.fillcolor(fillcolor)
    p.begin_fill()
    for i in range(2):
        p.forward(h)
        p.right(90)
        p.forward(w)
        p.right(90)
    p.end_fill()
    
p.right(90)
draw_rectangular(300,150)

p.goto(150,0)
draw_rectangular(300,150)

p.goto(0,300)
draw_rectangular(300,150)

p.goto(150,300)
draw_rectangular(300,150)

p.penup()
p.goto(0,0)
p.pendown()

p.penup()
p.goto(-50,175)
p.pendown()
draw_rectangular(100,50,"#EDF9F8",'#020C03')

p.penup()
p.goto(0,0)
p.pendown()

p.penup()
p.goto(100,175)
p.pendown()
draw_rectangular(100,50,"#EDF9F8",'#020C03')

p.penup()
p.goto(0,0)
p.pendown()

p.penup()
p.goto(100,-125)
p.pendown()
draw_rectangular(100,50,"#EDF9F8",'#020C03')


p.penup()
p.goto(0,0)
p.pendown()

p.penup()
p.goto(-50,-125)
p.pendown()
draw_rectangular(100,50,"#EDF9F8",'#020C03')

#vẽ nhà nhỏ bên cạnh
p.penup()
p.goto(0,0)
p.pendown()

p.penup()
p.goto(600,0)
p.pendown()
draw_rectangular(300,450)

p.penup()
p.goto(375,-100)
p.pendown()
draw_rectangular(200,100,"#EDF9F8",'#020C03')
p.penup()
p.goto(475,-100)
p.pendown()
draw_rectangular(200,100,"#D4C433",'#020C03')



turtle.done()
```
