```python
#mở turtle
import turtle
p=turtle.Turtle()
p.speed(30)
#chỉnh màu nền 
p.getscreen().bgcolor("#00FF00")

#chỉnh màu nền ngôi nhà
p.color('#000000')
p.fillcolor("#FFFF00")

#vẽ hình vuông
p.penup()
p.goto(-300,-100)
p.pendown()
p.begin_fill()
for i in range(4):
    p.forward(200)
    p.left(90)
p.end_fill()

#vẽ mái nhà
p.color('#000000')
p.fillcolor("#FFFF00")
p.penup()
p.goto(-300,100)
p.pendown()

p.begin_fill()
p.left(60)
p.forward(200)
p.left(240)
p.forward(200)
p.end_fill()

#vẽ hình trụ cho ngôi nhà
p.color('#000000')
p.fillcolor("#FFFF00")
p.begin_fill()
p.right(270)
for i in range(4):
    p.forward(200)
    p.left(90)   
p.end_fill()

#vẽ hình trụ cho ngôi nhà lần 2
p.color('#000000')
p.fillcolor("#FFFF00")

p.begin_fill()

p.forward(200)
p.right(120)

p.forward(200)
p.right(60)
p.forward(200)
p.right(120)
p.forward(200)
p.end_fill()

#cửa sổ của ngôi nhà
p.color('#000000')
p.fillcolor("#FFFF00")

p.penup()
p.goto(-150,-100)
p.pendown()
p.begin_fill()
for i in range(2):
    p.forward(150)
    p.left(90)
    p.forward(75)
    p.left(90)
p.end_fill()

p.color('#000000')
p.fillcolor("#FFFF00")

#cửa sổ còn lại của ngôi nhà

# mặt trời
p.color('#000000')
p.fillcolor("#FFFF00")

p.penup()
p.goto(600,250)
p.pendown()
p.begin_fill()
p.circle(50)
p.end_fill()

#vẽ cái cây
#thân cây
p.fillcolor('brown')
p.begin_fill()
p.penup()
p.goto(200, -130)
p.pendown()

p.forward(40)
p.left(90)
p.forward(20)
p.left(90)
p.forward(40)
p.left(90)
p.forward(20)
p.left(90)
p.end_fill()

#lá cây 1
p.penup()
p.goto(225, -90)
p.pendown()
p.right(90)
p.penup()
p.goto(225,-90)
p.pendown()
p.left(120)
p.fillcolor('lightgreen')
p.begin_fill()

p.forward(70)
p.left(120)
p.forward(70)
p.left(120)
p.forward(70)
p.end_fill()

#lá cây 2
p.penup()
p.goto(225, -30)
p.pendown()
p.left(120)
p.fillcolor('lightgreen')
p.begin_fill()

p.forward(70)
p.left(120)
p.forward(70)
p.left(120)
p.forward(70)
p.end_fill()

#lá cây 3
p.penup()
p.goto(225, 30)
p.pendown()
p.left(120)
p.fillcolor('lightgreen')
p.begin_fill()

p.forward(70)
p.left(120)
p.forward(70)
p.left(120)
p.forward(70)
p.end_fill()

turtle.done()
```
