```python
from tkinter import *
from tkinter import messagebox
windows = Tk()
windows.title('password manager')

website_lb = Label(windows, text = 'Website')
email_lb = Label(windows, text = 'Email/Username')
password_lb = Label(windows, text = 'Password')

website = Entry(windows)
email = Entry(windows)
password = Entry(windows)

list_info = []
def clear_entry_value():
    website.delete(0,END)
    email.delete(0,END)
    password.delete(0,END)

def add_password():
    global list_info # ch nh s dng list_info bn ngoi  dng trong hm
    print(f'Website: {website.get()}')
    print(f'Username: {email.get()}')
    print(f'Password: {password.get()}')
    info = (website.get(),email.get(),password.get())
    list_info.append(info)
    print(list_info)
    messagebox.showinfo("Add Password", "password added successfully")
    clear_entry_value()
    
    
def search_info():
    print(f"Looking on website: {website.get()}")
    for info in list_info:
        web_name = info[0]
        if web_name == website.get():
            username_text =info[1]
            password_text = info[2]
            messagebox.showinfo("Website information",f"Username: {username_text}, Password:{password_text}")
            break
            
search = Button(windows, text = 'Search', width = 10, command=search_info)
generate = Button(windows, text = 'Generate Password', width =20)
add = Button(windows, text = 'Add', width = 10, command=add_password)

website_lb.place(x = 100, y = 50)
email_lb.place(x = 100, y = 80)
password_lb.place(x = 100, y = 110)

website.place(x = 200, y = 50)
email.place(x = 200, y = 80)
password.place(x = 200, y = 110)

search.place(x = 350, y = 50)
generate.place(x = 350, y = 80)
add.place(x = 200, y = 150)

windows.mainloop()
```
