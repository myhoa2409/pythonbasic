```python
from tkinter import *
windows =Tk()
windows.title('Simple Calculator')

register_lbl = Label(windows, text = 'Register')
fullname_lbl = Label(windows, text = 'Full name')
address_lbl = Label(windows, text = 'Address')
email_lbl = Label(windows, text ='Email')

fullname_entry = Entry(windows)
address_entry = Entry(windows)
email_entry = Entry(windows)

btn_register = Button(windows, text = 'Register')
btn_login = Button(windows, text = 'login')

register_lbl.place(x=100, y=50)
fullname_lbl.place(x=10, y = 100)
address_lbl.place(x=10, y=150)
email_lbl.place(x=10, y= 200)

fullname_entry.place(x =100, y=100)
address_entry.place(x=100, y =150)
email_entry.place(x=100, y=200)

btn_register.place(x =100, y=250)
btn_login.place(x=160, y=250)

windows.mainloop()
```

```python
from tkinter import *
windows = Tk()
windows.geometry("400x300")
windows.eval('tk::PlaceWindow . center')

windows.title('Simple Calculator')




result_lb = Label(windows,text = 'Result')
num1 = Entry(windows)
num2 = Entry(windows)

def add():
    result = int(num1.get()) + int(num2.get())
    result_lb.config(text = result)

def subtract():
    result = int(num1.get()) - int(num2.get())
    result_lb.config(text = result)
    
def duplicate():
    result = int(num1.get()) * int(num2.get())
    result_lb.config(text = result)
    
def divide():
    result = int(num1.get()) / int(num2.get())
    result_lb.config(text = result)

add_bt = Button(windows, text = '+', width =5, command = add)
subtract_bt = Button(windows, text = '-', width =5, command = subtract)
duplicate_bt = Button(windows, text = 'x', width = 5, command = duplicate)
devide_bt = Button(windows, text = '/', width= 5, command = divide )

num1.pack()
num2.pack()
add_bt.pack()
subtract_bt.pack()
duplicate_bt.pack()
devide_bt.pack()

result_lb.pack()

windows.mainloop()
```

```python
from tkinter import *
windows = Tk()
windows.geometry("400x300")
windows.eval('tk::PlaceWindow . center')

windows.title('My info')

result_lb = Label(windows,text = 'My information')
firstname = Label(windows,text = 'First Name')
lastname = Label(windows,text = 'Last Name')
firstname_entry = Entry(windows)
lastname_entry = Entry(windows)

def showname():
    result = f'My Name is {firstname_entry.get() +" "+ lastname_entry.get()}'
    result_lb.config(text = result)


submit = Button(windows, text = 'Submit', width =5, command = showname)


result_lb.place(x=50,y=50)
firstname.place(x=50, y=80)
lastname.place(x=50, y=120)
firstname_entry.place(x=150, y=80)
lastname_entry.place(x=150, y=120)
submit.place(x=150,y=150)

result_lb.pack()

windows.mainloop()
```

```python
from tkinter import *
from tkinter import messagebox
windows = Tk()
windows.geometry("400x300")
windows.eval('tk::PlaceWindow . center')

windows.title('My BMI')



result_lb = Label(windows,text = 'BMI calculation')
myweight = Label(windows,text = 'My Weight (Kg)')
myheight = Label(windows,text = 'My height (cm)')
myweight_entry = Entry(windows)
myheight_entry = Entry(windows)

def BMI():
    result = f'Your BMI is {int(myweight_entry.get()) / ((int((myheight_entry.get()))/100)**2)}'
    messagebox.showinfo('Your BMI',f' your BMI is {result}')
    result_lb.config(text = result)
    

submit = Button(windows, text = 'Submit', width =5, command = BMI)


result_lb.place(x=50,y=50)
myweight.place(x=50, y=80)
myheight.place(x=50, y=120)
myweight_entry.place(x=150, y=80)
myheight_entry.place(x=150, y=120)
submit.place(x=150,y=150)

result_lb.place(x=50, y = 20)

windows.mainloop()
```
