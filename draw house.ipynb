```python
#mở turtle
import turtle
p=turtle.Turtle()

#chỉnh màu nền 
p.getscreen().bgcolor("#00FF00")

#chỉnh màu nền ngôi nhà
p.color('#000000')
p.fillcolor("#FFFF00")

#vẽ hình vuông
p.penup()
p.goto(-300,-100)
p.pendown()
p.begin_fill()
for i in range(4):
    p.forward(200)
    p.left(90)
p.end_fill()

#vẽ mái nhà
p.color('#000000')
p.fillcolor("#FFFF00")
p.penup()
p.goto(-300,100)
p.pendown()

p.begin_fill()
p.left(60)
p.forward(200)
p.left(240)
p.forward(200)
p.end_fill()

#vẽ hình trụ cho ngôi nhà
p.color('#000000')
p.fillcolor("#FFFF00")
p.begin_fill()
p.right(270)
for i in range(4):
    p.forward(200)
    p.left(90)   
p.end_fill()

#vẽ hình trụ cho ngôi nhà lần 2
p.color('#000000')
p.fillcolor("#FFFF00")

p.begin_fill()

p.forward(200)
p.right(120)

p.forward(200)
p.right(60)
p.forward(200)
p.right(120)
p.forward(200)
p.end_fill()

#cửa sổ của ngôi nhà
p.color('#000000')
p.fillcolor("#FFFF00")

p.penup()
p.goto(-150,-100)
p.pendown()
p.begin_fill()
for i in range(2):
    p.forward(150)
    p.left(90)
    p.forward(75)
    p.left(90)
p.end_fill()

p.color('#000000')
p.fillcolor("#FFFF00")

#cửa sổ còn lại của ngôi nhà

# mặt trời
p.color('#000000')
p.fillcolor("#FFFF00")

p.penup()
p.goto(600,250)
p.pendown()
p.begin_fill()
p.circle(50)
p.end_fill()

# tia mặt trời
p.penup()
p.goto(500,250)
p.pendown()
p.left(60)
p.forward(50)

#vẽ cái cây
p.color('#000000')
p.fillcolor("#FFFF00")
p.penup()
p.goto(400,0)
p.left(150)
p.pendown()
p.begin_fill()
p.circle(50,steps = 3)
p.end_fill()

p.color('#000000')
p.fillcolor("#FFFF00")
p.penup()
p.goto(390,-80)
p.pendown()
p.begin_fill()
p.circle(55,steps = 3)
p.end_fill()

p.color('#000000')
p.fillcolor("#FFFF00")
p.penup()
p.goto(375,-200)
p.pendown()
p.begin_fill()
p.circle(80,steps = 3)
p.end_fill()

#thân cây
p.color('#000000')
p.fillcolor("#FFFF00")
p.penup()
p.goto(410,-200)
p.right(70)
p.pendown()
p.begin_fill()
p.circle(30,steps = 4)
p.end_fill()


turtle.done()
```
