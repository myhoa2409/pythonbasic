```python
from tkinter import *
from tkinter import ttk
from tkinter.ttk import *
from tkinter import messagebox
from tkinter import filedialog
windows = Tk()
windows.title('HR manager')
windows.iconbitmap(r'D:\MYHOA\1. STUDY\1. PYTHON\PYF2202R1\pythonbasic')

#add some style
style = ttk.Style()
#pick a Theme
style.theme_use("default")

#configure the treeview color
style.configure("Treeview",
               background="#D3D3D3",
               foreground="black",
               rowheight=25,
               fieldbackground="#D3D3D3")

#change selectied color
style.map("Treeview",
         background=[("selected","#347083")])

#create treeview scrollbar
my_tree = ttk.Treeview(windows)
# create a scrollbar widget and set its command to the text widget
scrollbar = ttk.Scrollbar(windows, orient='vertical', command=my_tree.yview)
scrollbar.grid(row=9, column=10, sticky='ns')

#  communicate back to the scrollbar
my_tree['yscrollcommand'] = scrollbar.set



#define our columns
my_tree['columns'] = ("Code", "Name","Date Joined","basic salary","department",
                      "Working days","salary in month","Insurance","Tax","Net take home")

#formate our columns
my_tree.column("#0", width=80, minwidth=25)
my_tree.column("Code", anchor=CENTER, width=80)
my_tree.column("Name", anchor=W, width=120)
my_tree.column("Date Joined", anchor=W, width=100)
my_tree.column("basic salary", anchor=CENTER, width=80)
my_tree.column("department", anchor=W, width=100)
my_tree.column("Working days", anchor=CENTER, width=100)
my_tree.column("salary in month", anchor=CENTER, width=120)
my_tree.column("Insurance", anchor=CENTER, width=80)
my_tree.column("Tax", anchor=CENTER, width=80)
my_tree.column("Net take home", anchor=CENTER, width=120)

#create heading
my_tree.heading("#0", text ="Label", anchor = W)
my_tree.heading("Code", text ="Code", anchor = CENTER)
my_tree.heading("Name", text ="Name", anchor = W)
my_tree.heading("Date Joined", text ="Date Joined", anchor = W)
my_tree.heading("basic salary", text ="basic salary", anchor = CENTER)
my_tree.heading("department", text ="department", anchor = W)
my_tree.heading("Working days", text ="Working days", anchor = CENTER)
my_tree.heading("salary in month", text ="salary in month", anchor = CENTER)
my_tree.heading("Insurance", text ="Insurance", anchor = CENTER)
my_tree.heading("Tax", text ="Tax", anchor = CENTER)
my_tree.heading("Net take home", text ="Net take home", anchor = CENTER)

#add data
list_info = [('01','MY HOA','25/04/2022',15000,'ACCOUNTING'),
             ('02','NAM','19/04/2021',7000,'PRODUCTION'),
             ('03','THANH','19/10/2021',24000,'HR'),
             ('04','THANH TAM','25/04/2019',35000,'ACCOUNTING'),
             ('05','JACKSON','25/04/2021',9050,'QC'),
            ('06','TIMMY','29/04/2020',82000,'CUSTOM'),
            ('07','HUONG','30/04/2020',14000,'PRODUCTION'),
            ('08','TIM','01/04/2020',6800,'CUSTOM'),
            ('09','MY','09/04/2020',8700,'HR'),
            ('10','TOM','03/04/2020',35600,'ACCOUNTING'),
            ('11','TINH','06/04/2020',28900,'QC'),
            ('12','KHOA','05/04/2020',8060,'HR'),
            ('13','TIMMY','29/04/2020',9060,'CUSTOM'),
            ('14','KHUONG','10/04/2020',18080,'PRODUCTION'),
            ('15','HUYNH','23/04/2020',13800,'PRODUCTION')]

#create striped row tags
my_tree.tag_configure("oddrow",background="white")
my_tree.tag_configure("evenrow",background="lightblue")

#add data to table
global count
count=0

for record in list_info:
    if count % 2 ==0:
        my_tree.insert(parent="", index="end", iid=count, text="", 
                   values=(record[0],record[1],record[2],record[3],record[4]),tags=("evenrow",))
    else:
        my_tree.insert(parent="", index="end", iid=count, text="", 
                   values=(record[0],record[1],record[2],record[3],record[4]),tags=("oddrow",))
    count +=1


my_tree.grid(row=9, column=0, columnspan = 10)
        


# tạo label
#tạo font chữ kích cỡ chữ
HR = Label(windows, text = 'HR MANAGER',font=("Arial Bold", 50))

code = Label(windows, text = 'Code')
name = Label(windows, text = 'Name staffs')
department = Label(windows, text = 'Department')
datejoin = Label(windows, text = 'Date Join')
salary = Label(windows, text = 'Salary')

# tạo entry điền thông tin
code_et = Entry(windows)
name_et = Entry(windows)
date_et = Entry(windows)
salary_et = Entry(windows)

#combo box chọn phòng ban
combo = Combobox(windows, state="Department")
combo['value'] = ('accounting','HR','Production','custom','QA')

def clear_entry_value():
    code_et.delete(0,END)
    name_et.delete(0,END)
    salary_et.delete(0,END)
    date_et.delete(0,END)
    combo.delete(0,END)
    wh_et.delete(0,END)
#hàm add thêm nv
def add_staff():
    global list_info # chỉ định sử dụng list_info bên ngoài để dùng trong hàm
    global count
    if str(code_et.get()) == "" or str(name_et.get()) == "" or str(date_et.get()) == "" or str(salary_et.get()) == "":
        messagebox.showwarning("Error", "Please fill in all fields!")
    else:
        A = False
        for info in list_info:
            code_staff = info[0]
            if code_staff == code_et.get():
                A = True
                break
        if A == False:
            my_tree.insert(parent="", index="end", iid=count, text="",
                           values=(code_et.get(),name_et.get(),date_et.get(),salary_et.get(),combo.get(),))
            count += 1
            list_new = (code_et.get(),name_et.get(),date_et.get(),salary_et.get(),combo.get())
            list_info.append(list_new)
            print(list_info)
            clear_entry_value()
            messagebox.showinfo("Add Staffs information", "Added successfully!")
        else:
            messagebox.showerror("Error", "this code staff is already added!, pls add another code")
            clear_entry_value()
#code to remove all information
def remove_all():
    for record in my_tree.get_children():
        my_tree.delete(record)
#remove on selected
def remove_one():
    x = my_tree.selection()[0]
    my_tree.delete(x)

#remove many selection
def remove_many():
    x = my_tree.selection()
    for record in x:
        my_tree.delete(x)
    
    
#hàm tìm kiếm search
def search_staff():
    global list_info
    count = False
    if str(code_et.get()) == "":
        messagebox.showwarning("Error", "Please provide code!")
    else:
        for info in list_info:
            code_staff = info[0]
            if code_staff == code_et.get():
                count = True
                name = info[1]
                date = info[2]
                salary = info[3]
                dept = info[4]
                break
        if count == False:
            messagebox.showerror("Staffs information", "No saved this staff information !")
        else:
            messagebox.showinfo("Staffs information", 
                                f"Name: {name}. Salary: {salary}. Department: {dept} . Date Joined: {date}")


            
#add working hour
def add_total():
    for child in my_tree.get_children():
        code = my_tree.item(child,"values")[0]
        name = my_tree.item(child,"values")[1]
        date = my_tree.item(child,"values")[2]    
        salary = float(my_tree.item(child,"values")[3])
        dept = my_tree.item(child,"values")[4]
        wh = wh_et.get()
        my_tree.item(child,values=(code,name,date,salary,dept,wh))


#salary calculation def
def salary_cal():
    global add_total
    for child in my_tree.get_children():
        code = my_tree.item(child,"values")[0]
        name = my_tree.item(child,"values")[1]
        date = my_tree.item(child,"values")[2]    
        salary = float(my_tree.item(child,"values")[3])
        dept = my_tree.item(child,"values")[4]
        wh = float(my_tree.item(child,"values")[5])
        
        salary_month = round(salary * wh /24,0)
        ins = round(salary_month * 0.105,0)
            
        tntt = salary_month - ins - 11000
        if tntt<=0:
            pit = 0
        else:
            if tntt > 80000:
                pit = 0.35*tntt - 9850
            elif tntt<=80000 and tntt>52000:
                pit = 0.3*tntt -5850
            elif tntt<=52000 and tntt>32000:
                pit = 0.25*tntt-3250
            elif tntt<=3200 and tntt>18000:
                pit = 0.2*tntt-1650
            elif tntt<=18000 and tntt>10000:
                pit = 0.15*tntt-750
            elif tntt<=10000 and tntt>5000:
                pit = 0.1*tntt-250
            else:
                pit=0.05*tntt
        net = round(salary_month - ins - pit,2)
                
        #item gọi value add lên bảng   
        my_tree.item(child,values=(code,name,date,salary,dept,wh,salary_month,ins,round(pit,0),net))
    messagebox.showinfo("Salary staffs", f"Salary already presented")


#selected record
def select_record(e):#in advance
    #clear entry boxes
    clear_entry_value()
    
    #grab record number
    selected = my_tree.focus()
    
    #grab record number
    values = my_tree.item(selected,'values')
    
    #output to entry boxes
    code_et.insert(0,values[0])
    name_et.insert(0,values[1])
    salary_et.insert(0,values[2])
    date_et.insert(0,values[3])
    combo.insert(0,values[4])

#bind treeview
my_tree.bind("<ButtonRelease-1>", select_record)

#move row up
def up():
    rows = my_tree.selection()
    for row in rows:
        my_tree.move(row, my_tree.parent(row),my_tree.index(row)-1)

        
#move rows down
def down():
    rows = my_tree.selection()
    for row in reversed(rows):
        my_tree.move(row, my_tree.parent(row),my_tree.index(row)+1)

#update record update
def update_record():
    #grab the record number
    selected = my_tree.focus()
    #update record
    my_tree.item(selected,text="",
                 values=(code_et.get(),name_et.get(),date_et.get(),salary_et.get(),combo.get(),wh_et.get()))
    #clear entry 
    clear_entry_value()

# hàm add WH
def working_hour_add():
    #grab the record number
    selected = my_tree.focus()
    
    #update record
    my_tree.item(selected,text="",
                 values=(code_et.get(),name_et.get(),date_et.get(),salary_et.get(),combo.get(),wh_et.get(),))
    
    list_new = (code_et.get(),name_et.get(),date_et.get(),salary_et.get(),combo.get(),wh_et.get())
    list_info.append(list_new)
    print(list_info)
    
    #clear entry
    clear_entry_value()
#hàm tạo cửa sổ mới
def create():
    global my_text
    win = Toplevel(windows)
    my_text = Text(win, width=40, height=10,font=("Helvetica",16))
    my_text.pack(pady=20)
    
    open_text = Button(win,text="Open text File", command=open_txt)
    open_text.pack(pady=20)
    
    #create button 
    save_button = Button(win, text="save text", command = save_txt)
    save_button.pack(pady=20)
#open text file
def open_txt():
    global my_text
    text_file = filedialog.askopenfilename(initialdir=r"D:\MYHOA\1. STUDY\1. PYTHON\PYF2202R1\pythonbasic", title="Open text file",filetypes=(("Text File","*.txt"),))
    text_file = open(text_file,"r")
    stuff = text_file.read()
    
    my_text.insert(END,stuff)
    text_file.close()

#create save txt funtion
def save_txt():
    global my_text
    text_file = open("thuchanh.txt","w")
    text_file.write(my_text.get(1.0, END))
     
        
# tạo button thêm sửa tìm kiếm
liststaff = Button(windows, text = 'List staffs', command = create)
add = Button(windows, text = 'Add staffs', command = add_staff)
search = Button(windows, text = 'Search Staffs', command = search_staff)
edit = Button(windows, text = 'Remove all', command = remove_all)
remove_one = Button(windows, text="Remove one", command = remove_one)

# button tính lương, tính bảo hiểm, tính thuế
cal_salary = Button(windows, text = ' Salary calculation', command = salary_cal)

#thêm entry working_hours
wh = Label(windows,text="Working days")
wh.grid(row=1,column=4)
wh_et = Entry(windows)
wh_et.grid(row=1, column=5)

#thêm entry add working hour
wh_bt = Button(windows, text = "Add Working_days", command = add_total)
text = Label(windows, text="CALCULATION")
text.grid(row=2,column=0, columnspan=20)
#hiện label
code.grid(row=0, column=0)
name.grid(row=1, column=0)
salary.grid(row=0, column=4)
department.grid(row=1, column=2)
datejoin.grid(row=0, column=2)

#hiện entry
code_et.grid(row=0, column=1)
name_et.grid(row=1, column=1)
salary_et.grid(row=0, column=5)
date_et.grid(row=0, column=3)
#hiện combobox
combo.grid(row=1, column=3)

#hiện button
liststaff.grid(row=3, column=0)
add.grid(row=3, column=1)
search.grid(row=3, column=2)
edit.grid(row=3, column=3)
update = Button(windows, text="Update", command=update_record)
update.grid(row=3, column=4, columnspan = 1)
#hiện button tính lương
cal_salary.grid(row=3, column=7, columnspan = 1)
remove_one.grid(row=3, column =5)
wh_bt.grid(row=3,column=6)

#select button
select = Button(windows, text="Clear entry record", command=clear_entry_value)
select.grid(row=1,column=6)

move_up = Button(windows, text="Up",command=up)
move_up.grid(row=0,column=7)
move_down = Button(windows, text="down",command=down)
move_down.grid(row=1,column=7)

windows.mainloop()
```
